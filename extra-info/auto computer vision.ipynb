```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox
```

```python
img= cv2.imread("cars.jpg")
```

```python
plt.imshow(img)
```

```python
# 'box', 'label', ve 'count' değişkenlerine nesne algılama sonuçlarını atıyoruz.
# 'box' - Algılanan nesnelerin sınırlayıcı kutularının koordinatlarını içerir.
# 'label' - Algılanan her nesnenin sınıf etiketini içerir (örneğin: 'person', 'car').
# 'count' - Algılanan nesne sayısını içerir.
box, label, count = cvlib.detect_common_objects(img)
```

```python
output= draw_bbox(img,box,label,count)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output)
```
