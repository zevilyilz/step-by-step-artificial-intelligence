```python
import pandas as pd
import numpy as np
```

```python
r_columns = ["user_id", "movie_id", "rating"]
ratings = pd.read_csv("ml-100k/u.data", sep="\t", header=None, usecols=range(3), names=r_columns)
```

```python
r_cols=["movie_id", "title"]
movies= pd.read_csv("ml-100k/u.item", encoding="iso-8859-1", sep="|", header=None, usecols=range(2), names=r_cols )
```

```python
r_cols=["user_id", "Age", "Gender","Profession","Zipcode"]
user= pd.read_table("ml-100k/u.user", sep="|", names=r_cols)
```

```python
ratings= pd.merge(movies, ratings)
```

```python
ratings.title.value_counts()
```

```python
ratings.movie_id.nunique()
```

```python
movie_ratings= ratings.pivot_table(index=["user_id"],columns=["title"],values="rating")
# 'ratings' veri setini kullanarak bir pivot tablo oluşturuyoruz.
# index=["user_id"] -> Her bir satır bir kullanıcıyı temsil eder.
# columns=["title"] -> Her bir sütun bir filmi (başlığı) temsil eder.
# values="rating" -> Hücrelerde her kullanıcı ve film için verilen puanlar yer alır.
```

```python
movie_ratings
# matrix fact
```

```python
movie_ratings["Star Wars (1977)"]
```

```python
starwars_ratings= movie_ratings["Star Wars (1977)"]
```

```python
starwars_ratings.value_counts()
```

```python
movie_ratings[["101 Dalmatians (1996)", "Star Wars (1977)"]].corr()
```

```python
movie_ratings.corrwith(starwars_ratings)
# müsteri star wars izledikten sonra star wars ile en yükdrk korelasyona sahip film önerilir
```

```python
similar_movies= movie_ratings.corrwith(starwars_ratings)
```

```python
# similar_movies= similar_movies.dropna()
```

```python
similar_movies.sort_values(ascending=False)
```

```python
movie_stats= ratings.groupby("title").agg({"rating":[np.size,np.mean]})
```

```python
movie_stats.head()
```

```python
popular_movies= movie_stats[movie_stats["rating"]["size"]>100]
# 100 kişiden fazla izlenmiş filmler
```

```python
popular_movies.sort_values(("rating","mean"),ascending=False)
# ratingde ki meane göre küçükten büyüğe sıralandı
```

```python
# sıralanan filmlerin star warsla olan korelaasyonu
similar_movies_df= pd.DataFrame(similar_movies, columns=["similarity"])
popular_movies.columns= popular_movies.columns.get_level_values(0)
df= popular_movies.join(similar_movies_df)
```
