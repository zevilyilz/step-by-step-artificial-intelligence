# Language Detection 

Language detection is a key task in natural language processing (NLP), where the goal is to identify the language of a given text. While humans can easily identify familiar languages, machine learning models can accurately detect many languages. In this tutorial, we'll explore how to build a language detection model using Python and machine learning techniques.


```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
url ="https://raw.githubusercontent.com/amankharwal/Website-data/master/dataset.csv"
```

```python
df= pd.read_csv(url)
```

```python
df.head()
```

```python
# Count each language occurrence
language_counts = df["language"].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
language_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Language Counts in Dataset")
plt.xlabel("Language")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
x= np.array(df["Text"])
y= np.array(df["language"])
```

```python
tf_idf = TfidfVectorizer()
x = tf_idf.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=42)
```

```python
# logistic regression
l= LogisticRegression()
```

```python
l.fit(x_train,y_train)
```

```python
y_pred = l.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy:", accuracy)
```

### Conclusion

This project achieved 95.45% accuracy in language detection using TF-IDF vectorization with Logistic Regression, demonstrating effective language classification. The model is well-suited for multilingual systems, and further tuning could enhance performance even more.
