### Penn Treebank tokenization techniques
+ Word and Punctuation Separation: Separates punctuation from words. For example, "word." is tokenized into "word" and ".".

+ Handling Abbreviations: Applies special rules for correctly tokenizing abbreviations. For instance, "U.S." and "Dr." are tokenized properly.

+ Special Characters and Numbers: Applies specific rules for numbers and special characters. For example, "1234-5678" and "2.5" are tokenized correctly.

+ Separation and Contraction Rules: Handles contractions like "don't" and "can't" by separating them into "do" and "n't" or "can" and "n't".

+ Complex Syntactic Rules: Processes more complex structures and rules, ensuring accurate tokenization through both surface-level and deep syntactic rules.

### read the data 

```python
df= pd.read_csv("yelp (1).csv")
```

```python
df.head()
```

### convert everything to lowercase

```python
df["text"]= df["text"].str.lower()
```

### Remove punctuation and newline characters

```python
df["text"] = df["text"].str.replace(r"[^\w\s]", "", regex=True)  # Remove punctuation
df["text"] = df["text"].str.replace(r"\d+", "", regex=True)     # Remove numbers
df["text"] = df["text"].str.replace(r"\n", "", regex=True)      # Remove newlines
df["text"] = df["text"].str.replace(r"\r", "", regex=True)      # Remove carriage returns
```

```python
df["text"][0]
```

### sentiment analysis 
+ Sentiment analysis is a technique used to determine whether a piece of text expresses a positive, negative, or neutral sentiment.

```python
ybw= df[(df.stars==1)|(df.stars==5)]  # Filter rows with star ratings of 1 or 5
```

```python
ybw.head()
```

```python
ybw.reset_index(drop= True, inplace= True)  # Reset index and drop the old one
```

```python
ybw.head()
```

```python
x= ybw[["text"]]
y= ybw[["stars"]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
x.head()
```

```python
vect = CountVectorizer()
```

```python
x_new= vect.fit_transform(x["text"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=42, test_size=.20)
```

```python
nb= MultinomialNB()
```

```python
model= nb.fit(x_train,y_train)
```

```python
pred= model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```

```python
nltk.download("wordnet")
def stem_words(text):  # Define a function to extract word roots (lemmas)
    words = TextBlob(text).words  # Tokenize text into individual words
    return [word.stem() for word in words]  # Lemmatize each word (find its root form)
```

+ Stemming: Finds the most basic root form of a word (without considering grammatical meaning).
+ Lemmatization: Finds the linguistically correct root form of a word, considering its grammatical meaning.
