### Fetching Data from the Internet

```python
url= "https://en.wikipedia.org/wiki/Natural_language_processing"
```

```python
import requests
# for HTTP requests, HTTP forms the basis of data exchange on the web
```

```python
resp= requests.get(url)
```

```python
resp.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup= BeautifulSoup(resp.text, "html")   # Parse HTML content
soup
```

```python
for p in soup.find_all("p"):  # Find all <p> tags and print their text
    print(p.text)
```

```python
for a in soup.find_all("a"):  # for links
    print(a.text)
```
