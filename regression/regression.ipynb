# Featuare Engineering

```python
df.corr(numeric_only=True)
```

```python
abs(df.corr(numeric_only=True)["Price"].sort_values(ascending=False))
# mutlak dğer corrları
#  0.9 un altı 0.2 nin üstü
```

```python
x= df[["Cylinder", "Liter","Cruise","Mileage"]]
y= df[["Price"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
model= lr.fit(x_train, y_train)
```

```python
pred= model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mean_squared_error(y_test,pred)**0.5
```

```python
r2_score(y_test,pred)
```

```python
# lABEL encoding  
# rakama çevirme
```

```python
x= df[["Cylinder", "Liter","Cruise","Mileage","Make","Trim"]]
y= df[["Price"]]
```

```python
 x= pd.get_dummies(x,drop_first= True)  
# sayısala çevirdik
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model= lr.fit(x_train, y_train)
```

```python
pred= model.predict(x_test)
```

```python
mean_squared_error(y_test,pred)**0.5
# root mean squared error
```

```python
r2_score(y_test,pred)
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
```

```python
R= Ridge()
L= Lasso()
E= ElasticNet()
```

```python
Rmodel= R.fit(x_train,y_train)
```

```python
pred2= Rmodel.predict(x_test)
```

```python
mean_squared_error(y_test,pred2)**0.5
```

```python
r2_score(y_test,pred2)
```

```python
Lmodel= L.fit(x_train,y_train)
```

```python
pred3= Lmodel.predict(x_test)
```

```python
mean_squared_error(y_test,pred3)**0.5
```

```python
r2_score(y_test,pred3)
```

```python
Emodel= E.fit(x_train,y_train)
```

```python
pred4= Emodel.predict(x_test)
```

```python
mean_squared_error(y_test,pred4)**0.5
```

```python
r2_score(y_test,pred4)
```
